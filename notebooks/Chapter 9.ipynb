# Chpater 9 Priority Queues
## 9.1 The Priority Queue Abstract Data Type
### 9.1.1 Priorities
챕터 6에서 큐 ADT를 **first-in, first-out (FIFO)** 원칙에 따라 추가되고 삭제되는 객체들의 집합으로 소개한 적이 있었다. 먼저 전화를 건 손님을 먼저 응대하는 회사의 고객 콜 센터는 큐 ADT가 적용되는 가장 대표적인 사례일 것이다. 이러한 상황에서는 새로운 전화는 큐의 제일 뒤에 추가되고, 콜센터의 직원이 전화를 받을 수 있게 되면 큐의 제일 앞에 있는 사람이 그 직원과 연결된다.

실제 상황에서는 큐와 비슷한 구조를 이용해 객체를 관리하면서도 FIFO만으로는 충분하지 않은 그런 경우가 있다. 예를 들어 비행 교통 컨트롤 센터는 공항에 접근하는 많은 비행기 들 중에서 어떤 항공기가 착륙을 해도 좋을지 결정을 내려야 한다. 이 선택은 항공기가 얼마나 먼 거리를 날아왔는지, 얼마나 오래 기다렸는지, 연료가 얼마나 남았는지와 같은 요소를 고려해서 이루어진다. 이럴 때 FIFO에만 의존해서 착륙 결정을 내리는 건 적합하지 않다.

"먼저 온 사람을 먼저 접대하는" 정책이 적절해보이지만 그럼에도 다른 우선순위가 적용되는 상황들이 있다. 비행기의 예시를 다시 들자면, 어떤 비행기가 출발 한 시간 전에 예약이 꽉 찼다고 하자. 취소의 가능성으로 인해 항공사는 좌석을 얻기를 원하는 고객들의 대기 큐를 유지한다. 대기 승객의 우선순위는 그 승객의 체크인타임에도 영향을 받지만, 그 승객이 얼마나 항공사를 자주 이용하는지, 지불한 운임은 얼마인지와 같은 요소에도 영향을 받는다. 따라서 빈 자리가 났을 때 먼저 온 승객이 아니라 뒤늦게 온 승객이 자리를 받게 되는 경우가 있다.

이번 챕터에서는 **우선 순위 큐(priority queue)**로 알려진 새로운 추상 자료형을 소개한다. 우선순위 큐는 임의의 위치에 원소를 추가할 수 있고, 우선순위가 가장 높은 원소를 제거하는 것이 가능한, 우선순위가 매겨진 원소들의 집합이다. 원소가 우선순위 큐에 추가되면 유저는 그 원소와 관련된 **키(key)**를 제공함으로써 그 원소의 우선순위를 결정한다. *가장 작은* 키를 가진 원소가 큐에서 가장 먼저 제거될 원소가 된다(따라서, 키 1을 가진 원소가 키 2를 가진 원소보다 우선순위가 높다). 우선순위가 수치적으로 표현되는 것이 일반적이지만 `a < b`와 같이 키의 자연스러운 순서가 잘 정의되게끔 하는 일관적인 테스트를 지원하는 객체 타입이라면 모든 파이썬 객체가 키로 사용될 수 있다. 예를 들어 금융 아날리스트들은 특정한 자산들에 서로 다른 등급(rating, i.e.priorities)을 부여한다.

### 9.1.2 The Priority Queue ADT
형식적으로, 우리는 원소와 우선순위를 key-value 페어로 모형화한다. 우리는 우선순위 큐 ADT가 우선순위 큐 $P$에 대해 다음의 메소드를 지원하게끔 정의한다:
- **P.add(k, v):** 키 $k$와 값 $v$를 우선순위 큐 $P$에 추가한다.
- **P.min():** $P$에서 가장 작은 키를 가진 항목의 키와 값 튜플 (k,v)를 반환한다(항목을 제거하지는 않는다);우선순위 큐가 비어있다면 에러가 발생한다.
- **P.remove_min():** $P$에서 가장 작은 키를 가진 항목을 제거하고 키와 값 튜플 (k,v)를 반환한다. 우선순위 큐가 비어있다면 에러가 발생한다.
- **P.is_empty():** 우선순위 큐 $P$가 비어있다면 True를 반환한다.
- **len(P):** 우선순위 큐 $P$의 항목의 수를 반환한다.

우선순위 큐는 같은 키를 가진 여러 항목을 포함할 수도 있다. 이럴 때 `min`이나 `remove_min` 메소드는 가장 작은 키를 가진 항목 중 임의의 항목을 보고한다. 우선순위 큐에서 값(value)은 임의의 객체 타입이 될 수 있다. 우리의 우선순위 큐 초기 모델에서는 우선순위 큐에 한번 추가된 항목의 키는 고정된다고 가정하자. Section 9.5에서는 우선순위 큐 내 원소의 키를 업데이트하게끔 하는 확장을 살펴볼 것이다.

## 9.2 Implemetning a Priority Queue
이번 섹션에서는 위치 리스트 $L$(See Section 7.4)를 이용해서 우선순위 큐를 구현하는 법을 알아본다. $L$의 원소를 key로 정렬하는가의 여부에 따라 두 가지 구현 방식이 존재한다.

### 9.2.1 The Composition Design Pattern
우선순위 큐를 구현할 때 어려운 점은 항목들이 우리의 자료 구조 안에서 위치가 변하는 경우에 원소와 키 모두를 잘 추적하고 있어야 한다는 점이다. 이는 Section 7.6에서 각 원소의 카운트를 유지했던 경우와 비슷하다. 우리는 그러한 상황에서 **컴포지션 디자인 패턴(composition design pattern)**을 소개하고, `_Item` 클래스를 정의해서 각각의 인스턴스 안에 원소와 그 카운트가 함께 유지되도록 했었다.

우선순위 큐에서도 컴포지션을 이용해서 내부적으로 key $k$와 값 $v$로 이루어진 페어로서 항목을 저장할 수 있다. 우선순위 큐 구현에 이러한 개념을 구현하기 위해서 우리는 `PriorityQueueBase` 클래스 안에 중첩 클래스 `_Item`을 정의할 것이다. 우리는 key에 근거해서 item 인스턴스 $a$와 $v$를 비교할 수 있게끔 하는 `a < b` 문법 또한 정의할 것이다.

```python
class PriorityQueueBase:
    """Abstract base class for a priority queue."""
    
    class _Item:
        """Lightweight composite to store priority queue items."""
        __slots__ = '_key', '_value'
        
        def __init__(self, k, v):
            self._key = k
            self._value = v
            
        def __lt__(self, other):
            return self._key < other._key   # compare items b ased on their keys
        
    def is_empty(self):                  # concrete method assuming abstract len
        """Return True if the priority queue is empty."""
        return len(self) == 0
```

### 9.2.2 Implementation with an Unsorted List
우리의 첫번째 우선순위 큐의 구현에서는 원소들을 *정렬되지 않은* 리스트에 저장한다. `UnsortedPriorityQueue` 클래스는 `PriorityQueueBase` 클래스를 상속한다. 내부 저장소에서 key-value 페어는 상속받은 `_Item` 클래스의 인스턴스로서 composite으로 저장된다. 이 항목들은 클래스 안에서 `_data` 식별자를 갖는 `PositionalList`에 저장된다. 우리는 Section 7.4에서처럼 위치 리스트가 더블 링크드 리스트로 구현되어 ADT의 모든 연산이 $O(1)$ 시간 안에 이루어진다고 가정한다.

새로운 우선순위 큐가 생성되면 빈 리스트도 생성된다. 항상 리스트의 사이즈는 현재 우선순위 큐에 저장된 key-value pair의 수와 일치한다. 이러한 이유로 우리의 우선순위 큐 `__len__` 메소드는 단순히 내부 `_data` 리스트의 길이를 반환한다. 우리가 `PriorityQueueBase` 클래스에서 `is_empty` 메소드를 `__len__` 메소드의 호출에 의존하게끔 구체적으로 구현하였으므로 `is_empty` 메소드를 상속받아 이용할 수 있다.

`add` 메소드를 통해 우선순위 큐에 key-value 페어가 추가될 때마다 우리는 주어진 key와 value에 대한 `_Item` composite을 만들어서 리스트의 끝에 항목을 추가한다. 이러한 구현은 $O(1)$ 시간이 걸린다.

이제 남은 어려운 일은 `min`이나 `remove_min`이 호출될 때마다 가장 작은 키를 가진 항목의 위치를 찾는 것이다. 항목들이 정렬되어있지 않기 때문에 가장 작은 키를 가진 항목을 찾기 위해서는 모든 항목을 조사해야 한다. 편의를 위해, 가장 작은 키를 가진 항목의 *위치*를 반환하는 `_find_min` nonpublic 유틸리티 메소드를 정의하자. 위치를 알게 되면 `remove_min` 메소드를 통해 위치 리스트의 `delete` 메소드를 실행할 수 있다. `min` 메소드는 단순히 위치를 이용해 항목을 받아오기만 한다. 최소 키를 찾기 위한 루프로 인해 `min` `remove_min` 메소드는 모두 $O(n)$ 시간복잡도를 갖고, 이때 $n$은 우선순위 큐의 항목의 수이다.

```python
class UnsortedPriorityQueue(PriorityQueueBase): # base class defined _Item
    """A min-oriented priority queue implemented with an unsorted list."""
    
    def _find_min(self):                        # nonpublic utility
        """Return Position of item with minimum key."""
        if self.is_empty():                     # is_empty inherited from base class
            raise Empty('Priority queue is empty')
        small = self._data.first()
        walk = self._data.after(small)
        while walk is not None:
            if walk.element() < small.element():
                small = walk
            walk = self._data.after(walk)
        return small
    
    def __init__(self):
        """Create a new empty Priority Queue."""
        self._data = PositionalList()
        
    def __len__(self):
        """Return the number of items in the priority queue."""
        return len(self._data)
    
    def add(self, key, value):
        """Add a key-value pair."""
        self._data.add_last(self._Item(key,value))
        
    def min(self):
        """Return but do not remove (k,v) tuple with minimum key."""
        p = self._find_min()
        item = p.element()
        return (item._key, item._value)
    
    def remove_min(self):
        """Remove and return (k,v) tuple with minimum key."""
        p = self._find_min()
        item = self._data.delete(p)
        return (item._key, item._value)
```

### 9.2.3 Implementation with a Sorted List
우선순위 큐의 또 다른 구현 방식은 위치 리스트를 이용하면서도 항목들을 key를 이용해 정렬, 저장하는 방법이다. 이렇게 하면 리스트의 첫번째 원소는 key가 가장 작은 항목이 된다. 우리 `SortedPriorityQueue` 클래스는 리스트의 첫번째 원소가 가장 작은 키를 갖는다는 지식을 이용해서 `min`과 `remove_min`을 간단하게 구현한다. 우리는 위치 리스트의 `first` 메소드를 이용해서 첫번째 항목의 위치를 찾고 `delete` 메소드를 이용해 리스트에서 항목을 제거할 수 있다. 리스트가 더블 링크드 리스트로 구현되었다고 가정하면 `min`과 `remove_min`은 $O(1)$ 시간이 걸린다.

그러나 이러한 이득이 있으면 손실도 있는데, 이제 `add` 메소드를 실행하기 위해서 어느 위치에 새로운 항목을 넣어야할 지 결정해야 하게 되었다. 우리는 리스트의 끝에서 시작해서 자신보다 작은 key를 만날때까지 거꾸로 탐색하게끔 `SortedPriorityQueue`를 구현하였다. 최악의 경우에는 리스트의 처음까지 가야 하므로 $O(n)$ worst-case time을 갖는다. 요약하자면, 우선순위 큐를 구현하기 위해 정렬된 리스트를 이용할 경우 원소의 추가가 선형 시간에 이루어지지만 최소 키를 찾거나 제거하는 데에는 상수 시간이 걸린다.

<img width="400" alt="table-9.2" src="https://user-images.githubusercontent.com/20944657/36931168-ed60aaba-1ef3-11e8-8fc1-4a570aaf5c73.png">


```python
class SortedPriorityQueue(PriorityQueueBase):
    """A min-oriented priority queue implemented with a sorted list."""
    
    def __init__(self):
        """Create a new empty Priority Queue."""
        self._data = PositionalList()
        
    def __len__(self):
        """Return the number of items in the priority queue."""
        return len(self._data)
    
    def add(self, key, value):
        """Add a key-value pair."""
        newest = self._Item(key, value)             # make new item instance
        walk = self._data.last()     # walk backward looking for smaller key
        while walk is not None and newest < walk.element():
            walk = self._data.before(walk)
        if walk is None:
            self._data.add_first(newest)            # new key is smallest
        else:
            self._data.add_after(walk, newest)      # newest goes after walk
            
    def min(self):
        """Return but do not remove (k,v) tuple with minimum key"""
        if self.is_empty():
            raise Empty('Priority queue is empty.')
        p = self._data.first()
        item = p.element()
        return (item._key, item._value)
    
    def remove_min(self):
        """Remove and return (k,v) tuple with minimum key."""
        if self.is_empty():
            raise Empty('Priority queue is empty.')
        item = self._data.delete(self._data.first())
        return (item._key, item._value)
```

## 9.3 Heaps
지난 섹션에서 우선순위 큐 ADT를 구현하는 두 가지 전략에는 흥미로운 상충(trade-off)관계가 존재했다. *정렬이 되지 않은* 리스트를 이용해서 항목을 저장하면 최소 키를 가진 원소를 찾거나 제거하는 데에는 루프로 인해 $O(n)$ 시간이 필요했지만 추가에는 $O(1)$ 시간이면 충분했다. 그러나 *정렬된* 리스트를 이용할 경우에는 큐에 원소를 추가할 때 순서를 유지하기 위해 $O(n)$ 시간이 필요하지만 최소 key를 가진 원소를 찾거나 제거하는 데에는 $O(1)$ 시간이면 충분하다.

이번 섹션에서는 **이진 힙(binary heap)**이라고 알려진 자료 구조를 이용해서 더 효율적으로 우선순위 큐를 구현하는 방법을 알아본다. 이 자료 구조에서는 원소의 추가와 제거가 모두 로그 시간 안에 가능하므로 Section 9.2의 리스트-기반 구현보다 성능이 좋다. 힙이 이러한 성능의 향상을 이룰 수 있는 핵심적인 방법은 바로 이진 트리의 구조를 이용해서 정렬이 하나도 되지 않은 상태와 완전히 정렬이 된 상태 사이에서 절충안을 찾아내는 것이다.

### 9.3.1 The Heap Data Structure
<img width="600" alt="figure-9.1" src="https://user-images.githubusercontent.com/20944657/36931327-db1624ea-1ef6-11e8-828b-f94000961f8d.png">

힙은 각각의 위치들에 항목을 저장하면서 다음의 추가적인 두 성질을 만족하는 이진 트리 $T$이다:

**Heap-Order Property:** 힙 $T$에서는 루트를 제외한 위치 $p$에 저장된 key의 값이 $p$의 부모에 저장된 key의 값보다 크거나 같다.

heap-order 성질의 결과로, 루트에서 출발해서 $T$의 잎까지의 경로에서 만나게 되는 key들은 오름차순으로 정렬되어있다. 또한, 최소 키는 항상 $T$의 루트에 저장된다. 이러한 성질은 `min`이나 `remove_min`이 호출되었을 때 key가 가장 작은 항목을 찾기 쉽게 만들어주고, 이 때 최소 키 항목은 "힙의 가장 위(top)에 있다"라고 말한다. 참고로 알아둬야 할 것은 힙 자료 구조와 Section 15.1.1에서 다루는 메모리 힙과는 전혀 관련이 없다는 것이다.

효율성의 이유로 우리는 힙 $T$가 가능한 한 높이가 낮기를 원한다. 우리는 $T$가 **완전(complete)**하기를 요구함으로써 $T$의 높이를 낮게 유지한다.

**Complete Binary Tree Property:** $T$의 레벨 $0,1,2,...,h-1$이 가능한 한 많은 노드의 수를 갖고(레벨 $i$가 $2^{i}$개의 노드를 가질 때), 레벨 $h$의 나머지 노드들이 그 레벨에서 전부 왼쪽에 모여있을 때, 우리는 높이가 $h$인 힙 $T$를 **완전** 이진 트리라고 한다.

위 그림의 트리는 레벨 0,1,2가 꽉 차있고, 레벨 3의 6개 노드가 모두 왼쪽에 모여있으므로 완전 트리가 된다. 노드가 모두 왼쪽에 모여있다는 말을 엄밀하게 말하려면 이진 트리의 배열-기반 표현을 다룰 때 Section 8.3.2에서 **레벨 넘버링**을 다뤘던 것을 기억해야 한다. 이 레벨 넘버링을 이용하면, $n$개의 원소를 갖는 완전 이진 트리(complete binary tree)는 $0$부터 $n-1$까지의 레벨 넘버링을 갖는 위치로 이루어진 이진 트리라고 정의할 수 있다. 예를 들어, 위의 트리는 배열-기반의 표현방식을 이용했을 때 13개의 항목이 $A[0]$부터 $A[12]$까지 연속적으로 저장되므로 완전 이진 트리이다.

### The Height of a Heap
$T$의 높이를 $h$라 하자. $T$가 완전 트리여야한다고 주장하는 것은 다음의 중요한 결과를 불러온다.

**Proposition 9.2: ** $n$개의 원소를 저장하는 힙 $T$는 높이 $h = \lfloor logn \rfloor$를 갖는다.

**Justification: ** $T$가 완전하다는 사실로부터 우리는 레벨 $0$부터 $h-1$까지 노드의 수가 정확히 $1 + 2 + 4 + \cdots + 2^{h-1} = 2^{h} - 1$이고, 레벨 $h$의 노드 수는 $1$ 이상 $2^{h}$ 이하임을 알 수 있다. 따라서,<br>
$n \geq 2^{h}-1 + 1 = 2^{h}$ &nbsp;&nbsp;and&nbsp;&nbsp; $ n \leq 2^{h}-1 + 2^{h} = 2^{h+1} - 1$.

부등식 $2^{h} \leq n$의 양변에 로그를 씌우면 $h \leq logn$을 얻는다. 부등식 $n \leq 2^{h+1} - 1$을 정리해서 로그를 씌우면 $log(n+1) - 1 \leq h$를 얻는다. $h$가 정수이므로 이 두 부등식은 $h = \lfloor logn \rfloor$를 의미한다.

### 9.3.2 Implementing a Priority Queue with a Heap
만약 우리가 힙에 업데이트 연산을 실행하는데, 만약 그 시간이 트리의 높이에 비례하게만 할 수 있다면 명제 9.2에 따라 그 연산들은 로그 시간 안에 끝나게 된다. 이제 힙을 이용해서 다양한 우선순위 큐 메소드를 효율적으로 실행하는 방법을 알아보자. 우리는 Section 9.2.1에서 설명했던 컴포지션 패턴을 이용해서 key-value pair를 힙의 항목들로 저장할 것이다. `len`과 `is_empty` 메소드는 이전과 다를 것이 없고, `min` 연산은 항상 루트가 가장 키가 작은 원소를 갖고 있으므로 구현하는 것이 굉장히 쉽다. 그러나 `add`와 `remove_min` 메소드를 구현하기 위해서는 흥미로운 알고리즘들을 살펴봐야 한다.

### Adding an Item to the Heap
힙 $T$를 이용해 구현한 우선순위 큐에서 `add(k,v)`를 어떻게 실행할 수 있을지 생각해보자. 우리는 페어 $(k,v)$를 트리의 새로운 노드에 저장할 것이다. **완전 이진 트리 성질(complete binary tree property)**를 유지하려면 새로운 노드는 트리의 가장 낮은 레벨의 가장 오른쪽 위치에 저장되어야 하고, 만약 제일 낮은 층이 이미 꽉 차 있다면 새로운 레벨의 가장 왼쪽에 위치해야 한다.

### Up-Heap Bubbling After an Insertion
이렇게 노드를 저장하고 나면 트리 $T$는 완전(complete)하지만, **힙-순서 성질(heap-order property)**을 위반한다. 따라서, 위치 $p$가 $T$의 루트가 아닌 이상(우선순위 큐가 새로운 노드를 추가하기 전에 비어있었던 경우), 우리는 $p$의 키를 $p$의 부모 $q$의 키와 비교한다. 만약 $k_{p} \geq k_{q}$이면, 힙-순서 성질이 만족되고 알고리즘이 종료된다. 만약 $k_{p} \lt k_{q}$이면, 우리는 힙-순서 성질을 회복하기 위해 $p$와 $q$의 항목을 교체해야 한다. 이렇게 교체하면 새로운 항목이 한 레벨 위로 올라가게 된다. 이렇게 했는데도 여전히 힙-순서 성질이 위반된다면 이 과정을 계속해서 힙-순서 성질이 만족될 때까지 반복한다.

이렇게 새로운 항목을 계속해서 교체하며 위로 올리는 움직임을 관습적으로 **업-힙 버블링(up-heap bubbling)**이라 한다. 스왑(swap)은 힙-순서 성질의 위반을 해결하거나, 아니면 이 위반을 하나 더 높은 레벨로 전파한다. 최악의 경우, 업-힙 버블링은 새로운 항목이 힙 $T$의 루트까지 올라가게 한다. 따라서 최악의 경우 `add` 메소드에 의해 실행되는 스왑의 수는 $T$의 높이와 같다. 명제 9.2에 의해 bound는 $\lfloor logn \rfloor$이 된다.

### Removing the Item with Minimum Key
이제 우선순위 큐 ADT의 `remove_min` 메소드로 초점을 돌리자. 우리는 가장 작은 키를 가진 항목이 $T$의 루트 $r$에 저장되어 있음을 알고 있다(가장 작은 키를 가진 항목은 여러개일 수도 있지만 어쨌든 루트에 있는 항목은 최소 키를 가진 항목이다). 그러나, 일반적으로 우리는 단순히 노드 $r$을 지울 수 없는데, 두 서브트리가 연결되지 않은 채로 떨어져버리기 때문이다.
대신에 우리가 사용하는 방법은 힙의 모양이 **완전 이진 트리**를 유지하게 하는 방법이다. 루트를 제거하지 말고, $T$의 *마지막* 위치 $p$의 잎을 제거하고, 그 원소를 루트 $r$로 복사한다. 즉, $r$의 원소를 지우고 $p$의 원소로 대체하는 셈이다. 이 때 *마지막* 위치라 함은 가장 낮은 레벨의 가장 오른쪽 노드를 의미한다. 

### Down-Heap Bubbling After a Removal
$T$가 완전하긴 하지만 힙-순서 성질이 위배되므로 아직 끝나지 않았다. 만약 $T$가 한개의 노드(루트)만 갖고 있다면 힙-순서가 trivial하게 만족되므로 알고리즘이 끝난다. 그렇지 않다면 우리는 다음의 두 케이스를 구분해서 생각해야 한다. 처음에는 $p$는 $T$의 루트를 의미한다.
- 만약 $p$가 오른쪽 자식이 없다면 $c$를 $p$의 왼쪽 자식으로 둔다.
- 만약 $p$가 오른쪽 자식과 왼쪽 노드를 모두 갖는다면 $p$의 자식 중 키가 더 작은 자식을 $c$라 둔다.

만약 $k_{p} \leq k_{c}$이면 힙-순서 성질이 만족되고 알고리즘이 종료된다. 그렇지 않고 $k_{p} \gt k_{c}$이면 힙-순서 성질을 회복할 필요가 있다. 이를 위해서는 $p$와 $c$의 항목을 교체해야 한다. 이 때 $p$가 두 자식을 갖는다면 두 자식 중 더 작은 키를 가진 자식을 고른다는 데 주목할 필요가 있다. 이렇게 되면 $c$가 $p$보다 키가 작을 뿐 아니라, $c$의 형제자매보다도 더 작은 키를 갖게 된다. 그러면 $p$와 $c$의 원소가 교체되었을 때 힙-순서 성질이 국소적으로 만족된다. 만약 $p$의 두 자식 중 키가 더 큰 자식과 $p$가 교체되었다고 생각해보자. 이때 키가 더 큰 자식을 $c_{1}$, 키가 더 작은 자식을 $c_{2}$라 하자. 그러면 $p$와 $c_{1}$을 교체한 것이다. 이제 힙-순서 성질이 만족되는지 살펴보자. $c_{2}$는 $c_{1}$의 자식인데 $c_{2}$의 키가 $c_{1}$보다 크다. 따라서 힙-순서 성질이 만족되지 않게 된다. 이러한 이유로 키가 더 작은 자식과 $p$를 교체해야 하는 것이다.

$p$와 $c$를 교체했더라도 이제는 $c$에서 힙-순서 성질의 위반이 일어날 수 있다. 따라서 우리는 힙-순서 성질이 만족될 때까지 계속해서 알고리즘을 반복해서 트리 $T$의 아래로 스왑을 실행시킨다. 이러한 아래로의 스왑을 **다운-힙 버블링**이라고 한다. 스왑은 힙-순서 성질의 위배를 해결하거나, 힙의 한 레벨 아래로 이를 전파시킨다. 최악의 경우 항목이 힙의 가장 아래 레벨까지 내려갈 수 있다. 따라서 최악의 경우 `remove_min` 메소드에 의해 실행되는 스왑의 수는 $T$의 높이와 같다. 명제 9.2에 의해 bound는 $\lfloor logn \rfloor$이 된다.

### 9.3.3 Array-Based Representation of a Complete Binary Tree
이진 트리의 배열-기반 표현방식(Section 8.3.2)은 특히 완전 이진 트리 $T$에 적합하다. 배열 기반 표현방식의 구현에서 $T$의 원소들은 배열-기반 리스트 $A$에 저장되는데, 이 원소들은 다음의 인덱스로 $A$에 저장되었다:
- 만약 $p$가 $T$의 루트라면, $f(p) = 0$.
- 만약 $p$가 $q$의 왼쪽 자식이라면, $f(p) = 2f(q) + 1$.
- 만약 $p$가 $q$의 오른쪽 자식이라면, $f(p) = 2f(q) + 2$.

이러한 구현에서 $T$의 원소는 [0, n-1] 범위의 연속적인 인덱스를 갖고, $T$의 마지막 위치는 항상 인덱스 $n-1$을 갖는다.

<img width="600" alt="figure-9.4" src="https://user-images.githubusercontent.com/20944657/36934231-e569faec-1f29-11e8-9b1a-c393e65cdd34.png">

배열 기반의 힙 표현을 이용해서 우선순위 큐를 구현하면 노드-기반 트리 구조의 일부 복잡성을 피할 수 있게 된다. 특히, 우선순위 큐의 `add`와 `remove_min` 연산은 크기 $n$인 힙의 마지막 인덱스를 찾는 것에 의존하는데, 배열 기반의 구현방식에서 마지막 위치는 배열의 인덱스 $n-1$에 위치한다. 링크드 구조를 이용해서 구현된 완전 이진 트리에서 마지막 위치를 찾는 것은 배열 기반의 구현방식보다 더 많은 노력을 필요로 한다(연습문제 C-9.34 참고).

만약 우선순위 큐의 크기가 미리 알려져있지 않다면 배열 기반 표현을 사용할 경우 파이썬 `list`와 같이 배열을 동적으로 리사이징해야 할 필요가 생긴다. $n$개의 노드를 갖는 완전 이진 트리의 배열 기반 표현의 공간 복잡도는 $O(n)$이고, 원소를 추가하거나 제거하는 메소드의 time bound는 **amortized** bound가 된다.

### 9.3.4 Python Heap Implementation
우리는 배열 기반 표현방식을 이용할 것이므로 컴포지트 항목의 리스트를 둘 것이다. 비록 공식적으로 이진 트리 ADT를 이용하지는 않지만 아래의 코드는 자식 혹은 부모의 레벨 넘버링을 계산하는 많은 nonpublic 유틸리티 함수들을 포함하고 있다. 이 유틸리티 함수들은 *부모, 왼쪽, 오른쪽*과 같이 트리에서 주로 사용되는 용어들을 이용해서 알고리즘을 설명할 수 있게 해준다. 그러나 관련된 변수들은 "위치" 객체가 아닌 정수 인덱스이다. 우리는 재귀를 이용해서 `_upheap`과 `_downheap` 유틸리티의 반복을 구현한다.

```python
class HeapPriorityQueue(PriorityQueueBase): # base class defined _Item
    """A min-oriented priority queue implemented with a binary heap."""
    #--------------------- nonpublic behaviors -----------------------
    def _parent(self, j):
        return (j-1) // 2
    
    def _left(self ,j):
        return 2*j + 1
    
    def _right(self, j):
        return 2*j + 2
    
    def _has_left(self, j):
        return self._left(j) < len(self._data)    # index beyond end of list?
    
    def _has_right(self, j):
        return self._right(j) < len(self._data)   # index beyond end of list?
    
    def _swap(self, i, j):
        """Swap the elements at indices i and j of array."""
        self._data[i], self._data[j] = self._data[j], self._data[i]
        
    def _upheap(self, j):
        parent = self._parent(j)
        if j > 0 and self._data[j] < self._data[parent]:
            self._swap(j, parent)
            self._upheap(parent)                   # recur at position of parent
            
    def _downheap(self ,j):
        if self._has_left(j):
            left = self._left(j)
            small_child = left
            if self._has_right(j):
                right = self._right(j)
                if self._data[right] < self._data[left]:   # comparison is performed based on key
                    small_child = right     
            if self._data[small_child] < self._data[j]:
                self._swap(j, small_child)
                self._downheap(small_child)
                
    #---------------------- public behaviors ----------------------------
    def __init__(self):
        """Create a new empty Priority Queue."""
        self._data = []
        
    def __len__(self):
        """Return the number of items in the priority queue."""
        return len(self._data)
    
    def add(self, key, value):
        """Add a key-value pair to the priority queue."""
        self._data.append(self._Item(key, value))
        self._upheap(len(self._data) - 1)           # upheap newly added position
        
    def min(self):
        """Return but do not remove (k,v) tuple with minimum key.
        
        Raise Empty exception if empty.
        """
        if self.is_empty():
            raise Empty('Priority queue is empty')
        item = self._data[0]
        return (item._key, item._value)
    
    def remove_min(self):
        """Remove and return (k,v) tuple with minimum key.
        
        Raise Empty exception if empty.
        """
        if self.is_empty():
            raise Empty('Priority queue is empty.')
        self._swap(0, len(self._data) - 1)     # put minimum item at the end
        item = self._data.pop()                # and remove it from the list
        self._downheap(0)                      # then fix new root
        return (item._key, item._value)
```

- 최악의 경우, 업-힙 버블링과 다운-힙 버블링은 $T$의 높이와 같은 횟수의 스왑을 실시한다.

힙(heap) 자료 구조는 배열이나 링크 구조 중 어떤 것을 이용해서 구현하는가에 관계 없이 우선순위 큐 ADT를 구현하는 데에 매우 효율적이다. 힙을 이용해서 우선순위 큐 ADT를 구현하면 원소의 추가와 제거 모두 리스트를 이용할 때보다 빠르게 실행할 수 있게 된다.

### 9.3.6 Bottom-Up Heap Construction
이제 비어있는 힙에서 시작해서 `add` 연산을 $n$번 연속으로 실시하는 경우를 생각해보자. 그러면 최악의 경우 $nlogn$ 시간이 걸린다. 그러나 모든 $n$ key-value 페어가 이미 주어져있다고 생각해보자. 즉 모르는 key-value 페어들을 $n$번 연속 추가해서 힙을 만들어야 하는 경우가 아니라, 이미 주어진 $n$개의 key-value 페어를 힙으로 만들기만 하면 되는 문제를 다뤄보자는 것이다. 이경우에는 **bottom-up** 방식을 이용해서 $O(n)$ 시간 안에 $n$개의 원소를 힙으로 만들 수 있다. 따라서 이번 섹션에서는 bottom-up heap construction이 어떻게 이루어지는지 설명한다.

### Python Implementation of a Bottom-Up Heap Construction
이미 "down-heap" 유틸리티 함수를 정의했다면 bottom-up heap construction을 구현하는 건 매우 쉽다. $p$의 서브트리인 똑같은 사이즈의 두 힙을 "합치는(merging)" 것은 단순히 $p$의 항목을 다운-힙 해주기만 하면 된다. 예를 들어 위의 그림의 마지막 줄에서는 키 $14$를 다운-힙해서 두 힙을 합쳤다. 힙의 배열 기반 표현방식에서 만약 처음에 $n$개의 항목을 배열 내에 임의의 순서로 저장한다면, `_downheap`을 트리의 가장 *깊은* 레벨에서부터 루트의 순서로 호출하는 루프를 이용해서 bottom-up heap construction을 구현할 수 있다. 사실 루프는 가장 깊은 *nonleaf* 에서 시작해도 되는데, `_downheap`을 잎에 대해 실행할 경우 아무런 효과도 없기 때문이다.

이를 구현하기 위해서는 `HeapPriorityQueue` 클래스를 다음과 같이 수정해주면 된다.

```python
def __init__(self, contents=()):
    """Create a new priority queue.
    
    By Default, queue will be empty. If contents is given, it should be as an
    iterable sequence of (k,v) tuples specifying the initial contents.
    """
    self._data = [self._Item(k,v) for k,v in contents]          # empty by default
    if len(self._data) > 1:
        self._heapify()
        
def _heapify(self):
    start = self._parent(len(self) - 1)           # start at PARENT of last leaf
    for j in range(start, -1, -1):                # going to and including the root
        self._downheap(j)
```

### Asymptotic Analysis of Bottom-Up Heap Construction
Bottom-up heap construction은 $n$개의 키를 비어있는 힙에 추가하는 것보다 점근적으로 더 빠르다. 직관적으로, 우리는 모든 위치에 업-힙 연산을 실행하는 것이 아니라 다운-힙 연산을 실행하고 있는 셈이다. 트리 내에서는 위쪽 보다 아래쪽에 더 많은 노드가 있기 때문에 아래로 향하는 경로의 합은 선형 함수가 된다. 아래의 명제를 보자.

**Propostion 9.3:** 두 개의 키를 비교할 때 $O(1)$ 시간이 걸린다고 가정할 때, $n$개의 항목으로 이루어진 힙의 Bottom-up construction은 $O(n)$ 시간이 걸린다.

**Justification: ** 힙 생성의 가장 주된 비용은 모든 nonleaf 위치에서 실행되는 다운-힙 연산에 의해 발생한다. $\pi_{v}$를 위치 $v$에서 시작해서 $v$의 "중위 후계자(inorder successor)" 잎까지의 경로라 하자. 중위 운행법에서 위치 $p$의 왼쪽 서브트리를 먼저 방문하고 $p$를 방문한 후 다시 오른쪽 서브트리를 방문했던 것을 기억하는가? 그러면 중위 운행법에 따라 위치 $p$에서 출발할 때 가장 먼저 만나게 되는 잎까지의 경로를 생각해보자. $p$에서 시작한다는 것은 $p$의 왼쪽 서브트리는 이미 탐색을 마쳤다는 것이고, 이제 $p$의 오른쪽 서브트리로 가야 한다. 그런데 재귀적으로 생각해보면 $p$의 오른쪽 서브트리 안에서는 이제 또 왼쪽 서브트리부터 방문을 해야 한다. 이렇게 생각하면 $p$의 오른쪽 서브트리에서 가장 왼쪽에 있는 노드(잎)로의 경로가 우리가 찾는 경로가 될 것임을 알 수 있다. 이러한 경로를 $\pi_{v}$라 하는 것이다.

비록 $\pi_{v}$가 $v$에서 다운-힙 버블링을 실행했을 때의 경로는 아니지만, 그 길이인 $\lvert\lvert\pi_{v}\rvert\rvert$(엣지의 수)는 $v$를 루트로 하는 서브트리의 길이에 비례하고, 따라서 다운-힙 버블링 연산의 bound가 된다. 이를 이용하면, bottom-up heap construction 알고리즘의 전체 작동 시간 bound는 $\sum_{v} \lvert\lvert\pi_{v}\rvert\rvert$가 된다. 

우리가 보일 또 다른 주장은, 잎이 아닌 위치 $v$에 대해서 $\pi_{v}$들이 모두 edge-disjoint하다는 것이다. 그러면 경로 길이의 합은 트리의 총 엣지 수에 의해 bound되고, $O(n)$이 된다. 이를 보이기 위해서 "오른쪽으로 기울어진(right-leaning)" 엣지와 "왼쪽으로 기울어진(left-leaning)" 엣지를 구분해서 볼 필요가 있다. 트리에서 오른쪽으로 기울어진 엣지 $e$는 그 $e$가 $v$와 자식간의 관계를 나타내고 있을 때에만 $\pi_{v}$의 일부가 될 수 있다. 왼쪽으로 기울어진 엣지들은 잎에 도달할 때까지 계속 왼쪽으로 내려갔을 때 만나게 되는 잎에 따라 구분할 수 있다. nonleaf 노드들은 그 노드의 inorder successor를 만나게 해주는 left-leaning 엣지만 이용할 것이다. 그런데 각각의 nonleaf 노드들은 서로 다른 inorder successor를 가지므로, 어떠한 경로도 중복되는 left-leaning 엣지를 가질 수 없다. 따라서 bottom-up construction은 $O(n)$ 시간이 걸린다.

### 9.3.7 Python's heapq Module
파이썬의 표준 배포판에는 힙 기반의 우선순위 큐를 지원하는 `heapq` 모듈이 포함되어 있다. `heapq` 모듈은 우선순위 큐 클래스를 제공하지 않지만, 표준적인 파이썬 리스트가 힙으로서 관리될 수 있게끔 하는 함수들을 제공한다. 그 모델은 본질적으로 우리가 구현했던 모델과 같다. $n$개의 원소가 $L[0]$부터 $L[n-1]$까지 리스트 셀에 저장되어 있고, 가장 작은 원소가 루트 $L[0]$에 저장되게끔 레벨 넘버링이 되어있다. 그런데 `heapq`는 관련된 값들을 따로 저장해두지 않는다. 리스트의 원소들은 모두 key이다.

`heapq` 모듈은 리스트 $L$이 힙-순서 성질을 만족한다는 가정 하에 다음의 함수들을 지원한다:
- **heappush(L, e):** 리스트 $L$에 원소 $e$를 추가하고 힙-순서 성질을 유지한다. 이 함수는 $O(logn)$ 시간이 걸린다.
- **heappop(L):** 리스트 $L$에서 가장 작은 값을 가진 원소를 pop하고 반환한 후, 힙-순서 성질을 회복한다. 이 함수는 $O(logn)$ 시간이 걸린다.
- **heappushpop(L, e):** 리스트 $L$에 원소 $e$를 푸쉬하고 가장 작은 값을 pop하고 반환한다. 걸리는 시간은 $O(logn)$이지만 리스트의 크기가 변하지 않으므로 `pop`과 `push`를 따로 호출하는 것보다 효율적이다. 만약 새롭게 푸쉬된 원소가 가장 작은 값이라면 그 값이 바로 반환된다. 그렇지 않다면, 새로운 원소가 pop된 원소의 자리를 차지하고, down-heap이 실행된다.
- **heapreplace(L, e):** `heappushpop`과 비슷하지만, 푸쉬 전에 pop을 실행한다. 즉, 새로 추가된 원소는 pop의 반환값이 될 수 없다. 걸리는 시간은 $O(logn)$이지만 역시 따로 실행하는 것보다 효율적이다.

이 모듈은 이전에 힙-순서 성질을 만족하지 않았던 시퀀스에 대해서도 실행할 수 있는 추가적인 함수를 제공한다:
- **heapify(L):** 정렬되지 않은 리스트가 힙-순서 성질을 만족하게끔 변형시킨다. 이 함수는 bottom-up construction 알고리즘을 이용해서 $O(n)$ 시간 안에 실행된다.
- **nlargest(k, iterable):** `iterable`에서 가장 큰 $k$개의 값을 담은 리스트를 만든다. 이 함수는 $O(n + klogn)$ 시간이 걸린다. 이 때 $n$은 `iterable`의 길이이다 (see Exercise C-9.42).
- **nsmallest(k, iterable):** `iterable`에서 가장 작은 $k$개의 값을 담은 리스트를 만든다. 이 함수는 $O(n + klogn)$ 시간이 걸린다. 이 때 $n$은 `iterable`의 길이이다 (see Exercise C-9.42).

## 9.4 Sorting with a Priority Queue
우선순위 큐 ADT를 정의할 때 우리는 key로 어떠한 타입의 객체라도 이용할 수 있지만, 그 객체들은 서로 비교가 가능해서 key들의 집합이 자연스럽게 순서가 매겨질 수 있어야 한다고 했다. 파이썬에서는 그러한 순서를 정의하기 위해 주로 `<` 연산자에 의존하고, 이 `<` 연산자는 다음의 성질을 만족해야 한다:

- **Irreflexive property:** $k \nless k$.
- **Transitive property:** if $k_{1} \lt k_{2}$ and $k_{2} \lt k_{3}$, then $k_{1} \lt k_{3}$.

이러한 관계는 **strict weak order**라고 알려진 관계를 정의한다. strict weak order는 key들이 서로 같은 것을 허용하면서도, 전체적으로 봤을 때는 transitive property로 인해 **완전히 순서가 정해질 수 있게끔(totally orderded)** 한다. 우선순위 큐의 첫번째 응용사례로서 어떻게 집합 $C$를 정렬하는 데 쓰일 수 있는지 알아보자. 즉, 우리는 $C$의 원소들을 오름차순으로 정렬할 것이다. 이를 위한 알고리즘은 정말 단순하다. 모든 원소들을 비어있는 우선순위 큐에 넣고, `remove_min`을 계속 호출하기만 하면 된다.

파이썬 구현은 아래 코드와 같다. 여기서 우리는 $C$가 위치 리스트라고 가정했다. 따로 key가 있는 것이 아니므로 원소를 key와 value값으로 동시에 이용했다(e.g. `P.add(element, element)`).

```python
def pq_sort(C):
    """Sort a collection of elements stored in a positional list."""
    n = len(C)
    P = PriorityQueue()
    for j in range(n):
        element = C.delete(C.first())
        P.add(element, element)        # use element as key and value
    for j in range(n):
        (k,v) = P.remove_min()
        C.add_last(v)                  # store smallest remaining element in C
```

### 9.4.1 Selection-Sort and Insertion-Sort
우리의 `pq_sort` 함수는 우선순위 큐 클래스의 구현이 적절하기만 하다면 잘 작동할 것이다. 그러나 알고리즘의 전체 작동 시간은 주어진 우선순위 큐 클래스의 `add`, `remove_min` 연산의 작동 시간에 의존하게 된다. 이제 우리는 우선순위 큐 클래스를 잘 고르기만 하면 `pq_sort`가 몇몇 고전적인 정렬 알고리즘처럼 작동하게끔 할 수 있다는 것을 보일 것이다.

### Selection-Sort
만약 우리가 $P$를 정렬되지 않은 리스트로 구현한다면, `pq_sort`의 페이즈 1에서는 각각의 원소를 추가하는데 $O(1)$ 시간이 걸리므로 $O(n)$ 시간이 걸릴 것이다. 페이즈 2에서는 `remove_min` 연산 시간이 $P$의 사이즈에 비례하고, 계산의 병목은 페이즈 2에서 최소 원소를 반복적으로 "선택"하는 과정에서 발생한다. 이러한 이유로 이 알고리즘은 **selection-sort**라고 알려져 있다.

$P$의 사이즈는 `remove_min`을 실행할 때마다 $n$부터 1씩 감소해서 0이 된다. 따라서 첫번쨰 연산은 $O(n)$이 걸리고, 두번째는 $O(n-1)$, ...이 된다. 따라서 두번째 페이즈에 필요한 총 시간은 $O(n + (n-1) + \cdots + 2 + 1) = O(\sum_{i=1}^{n} i)$가 된다. 그런데 $\sum^{n}_{i=1} i = n(n+1)/2$ 이므로, 페이즈 2는 전체 selection-sort 알고리즘 시간에 필요한 $O(n^{2})$ 시간을 필요로 한다.

### Insertion-Sort
만약 우리가 $P$를 정렬된(sorted) 리스트로 구현한다면 `remove_min` 연산에 필요한 시간이 $O(1)$이 되므로 페이즈 2의 작동 시간을 $O(n)$으로 줄일 수 있다. 그러나 불행히도 이제 페이즈 1이 작동 시간의 병목이 된다. 최악의 경우 `add` 연산은 현재 $P$ 사이즈에 비례한다. 이러한 정렬 알고리즘은 **insertion-sort**라고 알려져 있다. 사실 우선순위 큐에 원소를 추가하는 우리의 구현 방식은 Section 7.5에서 다뤘던 insertion-sort의 과정과 거의 유사하다.

페이즈 1 insertion-sort의 worst-case 작동 시간은<br> $O(1 + 2 + ... + (n-1) + n) = O(\sum^{n}_{i=1}i)$이다. 따라서 페이즈 1의 시간 복잡도는 insertion-sort와 마찬가지로 $O(n^{2})$가 된다. 그러나 selection-sort와의 차이점은, insertion-sort가 *best-case* 작동 시간 $O(n)$을 갖는다는 것이다.

### 9.4.2 Heap-Sort
이미 봤듯이, 우선순위 큐를 힙으로 구현하면 우선순위 큐 ADT의 모든 메소드를 로그 시간이나 그것보다 더 빠르게 실행할 수 있게 된다. 따라서 모든 우선순위 큐 메소드에 대한 빠른 작동 시간이 필요한 상황에는 힙을 이용하는 것이 적합하다. 이제 우선순위 큐를 힙으로 구현했을 때 `pq_sort`의 작동 시간을 살펴보자.

페이즈 1에서 $i^{th}$ `add` 연산은 $O(logi)$ 시간을 필요로 하는데, `add` 연산 이후에 힙 안에 $i$개의 원소가 존재하기 때문이다. 따라서 페이즈 1은 $O(nlogn)$ 시간을 필요로 한다.(Section 9.3.6에서 다룬 bottom-up heap construction을 이용하면 $O(n)$으로 줄일 수 있다.)

페이즈 2에서 $j^{th}$ `remove_min` 연산은 $O(log(n-j+1))$ 시간을 필요로 하는데, 연산이 실행될 때 힙 안에 $n-j+1$개의 원소가 있기 때문이다. 이제 $j$에 대해 합하면 페이즈 2는 $O(nlogn)$ 시간을 필요로 한다. 따라서 전체 우선순위 큐 정렬 알고리즘은 $O(nlogn)$ 시간 복잡도를 갖는다. 이 정렬 알고리즘은 **heap-sort**라고 더 잘 알려져있고, 그 성능은 아래의 명제에 잘 요약되어있다.

**Proposition 9.4:** heap-sort 알고리즘은 $C$의 두 원소가 $O(1)$ 시간 안에 비교될 수 있다는 가정 하에, $n$개의 원소를 갖는 집합 $C$를 $O(nlogn)$ 시간 안에 정렬할 수 있다.

### Implementing Heap-Sort In-Place

만약 $C$가 파이썬 리스트와 같은 배열 기반 시퀀스로 구현되었다면 heap-sort의 시간을 빠르게 할 수 있을 뿐 아니라, 보조용의 힙 자료 구조를 직접 만드는 대신 리스트의 일부분에 힙을 저장하게 함으로써 공간 복잡도를 상수배만큼 줄이는 것이 가능하다. 이는 알고리즘을 다음과 같이 수정함으로써 가능하다:
1. 힙 연산을 *maximum-oriented* 힙이 되게끔 다시 정의한다. 즉, 모든 위치의 key는 자식의 key보다 크거나 같아야 한다. 이는 알고리즘을 수정하거나, key의 순서를 반대로 정의하는 방식을 통해 구현할 수 있다. 이 알고리즘을 실행하는 동안 우리는 리스트의 왼쪽 부분(인덱스 $i-1$까지)을 힙의 원소를 저장하는데 쓸 것이고, 리스트의 오른쪽 부분(인덱스 $i$부터 $n-1$까지)을 시퀀스의 원소를 저장하는 데 쓸 것이다. 따라서 $C$의 처음 $i$번째 원소들은 힙의 배열-기반 표현을 의미하게 된다.

2. 알고리즘의 페이즈 1에서 힙과 시퀀스의 경계를 오른쪽으로 한 칸씩 옮긴다. 즉, $i^{th}$ 스텝에서 우리는 $i-1$ 인덱스에 원소를 추가하여 힙을 확장한다.

3. 알고리즘의 페이즈 2에서 힙과 시퀀스의 경계를 왼쪽으로 한 칸씩 옮긴다. 즉, $i^{th}$ 스텝에서 우리는 힙의 가장 큰 원소를 제거하고 $n-1$ 인덱스에 원소를 저장한다.

일반적으로 우리는 정렬 알고리즘이 원소를 저장하고 있는 시퀀스 외에 적은 양의 메모리만 이용할 경우 그 정렬 알고리즘을 **in-place**라고 부른다. 위의 변형된 heap-sort 알고리즘은 시퀀스 밖으로 원소를 옮겼다가 다시 넣는 것이 아니라 순서만 바꿀 뿐이므로 in-place 알고리즘이라 불릴만한 자격이 있다. 아래의 그림은 in-place heap-sort의 페이즈 2를 그림으로 나타낸 것이다.

## 9.5 Adaptable Priority Queues
우선순위 큐 ADT의 메소드는 정렬과 같은 기본적인 응용 사례를 위한 목적이라면 충분하다. 그러나 아래와 같은 상황에서는 추가적인 메소드가 필요할 수도 있다.
- 비관적인 승객이 기다리는 것에 지쳐서 탑승 시간 전에 먼저 떠나기로 결정했고, 대기 리스트에서 본인을 삭제해줄 것을 요구했다. 이런 경우 우선 순위에서 해당 승객의 항목을 지워야 하지만 항상 그 승객이 최우선순위를 갖고 있을 것이라는 보장이 없으므로 `remove_min` 만으로는 이러한 경우를 처리할 수가 없다. 따라서 우리는 임의의 항목을 지울 수 있는 `remove` 메소드가 필요하다.

- 또 다른 승객은 기다리던 와중에 자신의 골드 회원 카드를 발견해서 직원에게 이 카드를 보여줬다. 이 경우 그녀의 우선순위가 골드 회원에 맞게 수정되어야 하는데, 이러한 수정을 위한 `update` 메소드가 필요하다.

이번 섹션에서는 **적응가능한 우선순위 큐(adaptable priority queue)** ADT를 설명하고 이 ADT를 어떻게 힙 기반 우선순위 큐의 확장 버전으로서 구현할 수 있는지 알아본다.

### 9.5.1 Locators
`update`와 `remove` 메소드를 효율적으로 구현하기 위해서는 전체 집합에 대한 선형 탐색을 하지 않고서도 사용자의 원소를 찾는 메커니즘이 필요하다. 이러한 목적을 위해서 우리는 새로운 원소가 우선순위 큐에 구현될 때마다 **locator**라고 알려진 특별한 객체를 사용자에게 반환하게끔 한다. 이제 `update`와 `remove` 메소드를 실행하려는 사람은 다음과 같이 적절한 locator를 파라미터로 제공해야 한다:
- **P.update(loc, k, v):** locator `loc`에 의해 식별되는 항목의 key와 value를 교체한다.
- **P.remove(loc):** locator `loc`에 의해 식별되는 항목을 우선순위 큐에서 제거하고 그 (key, value) 페어를 반환한다.

이 locator 추상화는 Section 7.4의 위치 리스트와 Chapter 8의 트리에서 이용했던 `Position` 추상화와 유사한 것처럼 보인다. 그러나 locator는 위치(Position)과 달리 그 구조 내에서 원소가 위치하는 실제의 공간을 나타내지 않는다. 우선순위 큐에서 임의의 원소는 그 원소와 직접 관련이 있어 보이지 않는 연산에 의해서도 계속해서 위치가 변할 수 있다. 그럼에도 불구하고 큐 내의 어디엔가 그 항목이 존재하기만 한다면 locator는 여전히 유효하다.

이 때 주목해야 할 점은 locator 인스턴스들이 자기 자신의 정체성까지 바꾸지는 않는다는 것이다. 예를 들어 `token`은 여전히 같은 인스턴스를 참조하고 있다. 그저 인스턴스 안의 세번째 필드값이 변경됐을 뿐이다.

이러한 표현방식을 이용하면 적응가능한 우선순위 큐 ADT를 구현하는 것은 간단하다. `update`나 `remove`의 파라미터로 locator 인스턴스가 전달되면 힙 내에서 그 원소가 어디에 있는지 쉽게 알 수 있다. 이러한 지식이 있으면 key의 업데이트를 위해서 그저 업-힙 버블링이나 다운-힙 버블링을 실행하기만 하면 된다. 만약 임의의 원소를 제거하고 싶다면 마지막 위치의 원소를 locator 인스턴스가 가리키는 원소로 이동시키고 적절한 버블링을 실행해주기만 하면 된다.

### Python Implementation
여기까지의 설명을 따라왔다면 코드만 보고도 이해가 될 것이다. 따라서 자세한 설명은 생략한다.

```python
class AaaptableHeapPriorityQueue(HeapPriorityQueue):
    """A locator-based priority queue implemented with a binary heap."""
    
    #------------------- nested Locator class ------------------------
    class Locator(HeapPriorityQueue._Item):
        """Token for locating an entry of the priority queue."""
        __slots__ = '_index'                     # add index as additional field
        
        def __init__(self, k, v, j):
            super().__init__(k,v)
            self._index = j
            
    #------------------ nonpublic behaviors --------------------------
    # override swap to record new indices
    def _swap(self, i, j):
        super()._swap(i,j)                       # perform the swap
        self._data[i]._index = i                 # reset locator index (post-swap)
        self._data[j]._index = j                 # reset locator index (post-swap)
        
    def _bubble(self, j):
        if j > 0 and self._data[j] < self._data[self._parent(j)]:
            self._upheap(j)
        else:
            self._downheap(j)
            
    def add(self, key, value):
        """Add a key-value pair."""
        token = self.Locator(key, value, len(self._data)) # initialize locator index
        self._data.append(token)
        self._upheap(len(self._data) - 1)
        return token
    
    def update(self, loc, newkey, newval):
        """Update the key and value for the entry identified by Locator loc."""
        j = loc._index
        if not (0 <= j < len(self) and self._data[j] is loc):
            raise ValueError('invalid locator')
        loc._key = newkey
        loc._value = newval
        self._bubble(j)
        
    def remove(self, loc):
        """Remove and return the (k,v) pair identified by Locator loc."""
        j = loc._index
        if not (0 <= j < len(self) and self._data[j] is loc):
            raise ValueError('Invalid locator')
        if j == len(self) - 1:                 # item at last position
            self._data.pop()                   # just remove it
        else:
            self._swap(j, len(self)-1)         # swap item to the last position
            self._data.pop()                   # remove it from the list
            self._bubble(j)                    # fix item displaced by the swap
        return (loc._key, loc._value)
```
